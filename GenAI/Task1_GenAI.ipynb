### **Task 4: ChatPromptTemplate System (15%)**

**Input**

  role = 'teacher'

  topic = 'Neural Networks'

```python
role=input("Enter role: ")
topic=input("Enter topic: ")

if role=="teacher":
  chat_template = ChatPromptTemplate.from_messages([
  ("system","You are acting as a {role}"),
  ("user","Explain {topic} clearly step by step")
  ])

elif role=="interviewer":
  chat_template = ChatPromptTemplate.from_messages([
      ("system","You are acting as a {role}"),
      ("user","Ask 3 questions about {topic}")
  ])

elif role=="motivator":
  chat_template = ChatPromptTemplate.from_messages([
      ("system","You are acting as a {role}"),
      ("user","Explain {topic} as a story")
  ])

messages = chat_template.format_messages(role=role,topic=topic)

print("\nGenerated Chat Messages:")
print(messages)
```

### **Task 5: Input Validation Layer (10%)**

**Create function:**

    validate_inputs(audience, tone)




```python
def validate_inputs(audience, tone):
    valid_audience=["beginner", "intermediate", "expert"]
    valid_tone=["formal", "casual", "fun"]

    if audience not in valid_audience:
        print("\nInvalid audience. Assigning default to beginner")
        audience="beginner"

    if tone not in valid_tone:
        print("Invalid tone. Assigning default to formal")
        tone="formal"

    return audience, tone

a=input("Enter audience: ")
t=input("Enter tone: ")
audience,tone=validate_inputs(a,t)
print("\nfDefault Audience:",audience,"\nDefault Tone:",tone)
```

### **Task 6: Prompt Generator App (15%)**

**Create function:**

    generate_prompt(topic, audience, tone, style)



```python
def generate_prompt(topic, audience, tone, style):

    audience, tone = validate_inputs(audience, tone)

    templates = {
        "teaching": PromptTemplate(
            input_variables=["topic","audience","tone"],
            template="Explain {topic} for {audience} in a {tone} teaching style"
        ),

        "interview": PromptTemplate(
            input_variables=["topic","audience","tone"],
            template="Create 3 interview questions about {topic} for {audience} in a {tone} tone"
        ),

        "storytelling": PromptTemplate(
            input_variables=["topic","audience","tone"],
            template="Explain {topic} for {audience} in a {tone} storytelling style"
        )
    }

    prompt = templates[style].format(
        topic=topic,
        audience=audience,
        tone=tone
    )

    return prompt

print("Input Values")
print("Topic: Neural Networks")
print("Audience: beginner")
print("Tone: fun")
print("Style: storytelling\n")

# Example
print(generate_prompt("Neural Networks", "beginner", "fun", "storytelling"))
```
